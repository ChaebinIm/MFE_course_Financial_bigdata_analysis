```python
raw_data = pd.read_csv('./kospi_1분봉_201910_202010.csv')
tech = pd.read_csv('./tech.csv')
```

```python
raw_data['datetime'] = raw_data['일자'] + ' ' + raw_data['시간']
raw_data['datetime'] = pd.to_datetime(raw_data['datetime'], format='%Y/%m/%d')
```

```python
data = raw_data[['datetime', '종가']]
data.columns = ['datetime', 'close']
data = pd.concat([data, tech], axis = 1)
original_col = data.columns.to_list()
```

```python
data.head()
```

```python
"""
    해야 되는 것.
    1. (1~20)분 후의 가격 ~ price_1m
    2. (1~20)분 후의 로그수익률 ~ lgret_1m
    3. (1~20)분 후의 절대수익률 ~ ret_1m
"""
for i in range(1, 21):
    num_label = str(i) + 'm'
    data['price_' + num_label] = data['close'] - data['close'].iloc[i:].reset_index(drop = True)
    data['ret_' + num_label] = (data['close'].iloc[i:].reset_index(drop = True) - data['close'])/data['close']
    data['logret_' + num_label] = np.log(data['close'].iloc[i:].reset_index(drop = True)/data['close'])
```

```python
def extract_data(method, period): # method : ('price', 'ret', 'logret'), period = 1~20
    label = method + '_' + str(period) + 'm'
    cols = original_col.copy()
    cols.append(label)
    _data = data[cols]
    _data.set_index('datetime', inplace = True, drop = True)
    return _data
```

```python
# 120분 평균과 분산
mean_120m = data['close'].rolling(120).mean()
std_120m = data['close'].rolling(120).std()

# y date list
date_list = [1, 3, 5, 10, 15, 20]
# making y
for i in date_list:
    close_after = data['close'].iloc[i:].reset_index(drop = True)
    nans = np.empty(i)
    nans = np.nan
    close_after = close_after.append(pd.Series(nans, index = np.arange(len(close_after), len(close_after) + i)))
    label =  'y_' + str(i) +'m'
    lower_bound = mean_120m - std_120m / np.sqrt(120//i)
    upper_bound = mean_120m + std_120m / np.sqrt(120//i)
    # up & down 만들기 전에, nan을 없애야 함! nan은 false로 간주된다.
    up = close_after >= upper_bound
    down = close_after < lower_bound
    y = 1 * up -1 * down
    data[label] = y
```
